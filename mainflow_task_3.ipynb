Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load the dataset


```python
df = pd.read_csv('tips.csv')
```

print the top 10 rows

```python
print(df.head(10))
```

Example 1: Bar chart - Total tip by day

```python
plt.figure(figsize=(10, 6))
df.groupby('day')['tip'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Tip by Day')
plt.xlabel('Day')
plt.ylabel('Total Tip')
plt.show()
```

Example 2: Line chart - Average tip by total bill

```python
plt.figure(figsize=(10, 6))
df.groupby('total_bill')['tip'].mean().plot(kind='line', marker='o', color='green')
plt.title('Average Tip by Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Average Tip')
plt.grid(True)
plt.show()
```

Additonal Exercises

Scatter Plot: Total Bill vs. Tip, colored by Day

```python
plt.figure(figsize=(10, 6))
colors = {'Thur': 'red', 'Fri': 'blue', 'Sat': 'green', 'Sun': 'purple'}
plt.scatter(df['total_bill'], df['tip'], c=df['day'].apply(lambda x: colors[x]), alpha=0.5)
plt.title('Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(colors.keys())
plt.show()
```

Bar Chart: Average Tip by Day

```python
plt.figure(figsize=(10, 6))
df.groupby('day')['tip'].mean().plot(kind='bar', color='orange')
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.show()
```

Histogram: Distribution of Total Bill

```python
plt.figure(figsize=(10, 6))
df['total_bill'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
